## Introduction
This notebook performs an Exploratory Data Analysis (EDA) on a credit card transactions dataset.

The goal is to understand the data structure, detect patterns related to fraudulent transactions, and highlight challenges such as severe class imbalance, which is typical in fraud detection problems.

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
```

## Basic Dataset Overview

```python
df = pd.read_csv('../data/creditcard.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

## Target Variable Analysis (Fraud vs Non-Fraud)

```python
df['Class'].value_counts()
```

```python
df['Class'].value_counts(normalize=True) * 100
```

## Class Imbalance

The dataset is highly imbalanced:
- Class 0: Legitimate transactions
- Class 1: Fraudulent transactions

Fraud cases represent a very small percentage of the total transactions, making accuracy an unreliable evaluation metric for this problem.


## Visualization of Class Imbalance

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Class', data=df)
plt.title('Class Distribution')
plt.xlabel('Transaction Class')
plt.ylabel('Count')
plt.show()
```

## Transaction Amount Analysis

```python
df['Amount'].describe()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Amount'], bins=50, kde=True)
plt.title('Transaction Amount Distribution')
plt.xlabel('Amount')
plt.show()
```

## Amount by Class (Fraud vs Non-Fraud)

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Class', y='Amount', data=df)
plt.title('Transaction Amount by Class')
plt.xlabel('Class')
plt.ylabel('Amount')
plt.show()
```

## Time Feature Analysis

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Time'], bins=50)
plt.title('Transaction Time Distribution')
plt.xlabel('Time')
plt.show()
```

## Correlation Overview

```python
plt.figure(figsize=(12,8))
corr = df.corr()

sns.heatmap(
    corr,
    cmap='coolwarm',
    center=0,
    linewidths=0.5
)

plt.title('Feature Correlation Heatmap')
plt.show()
```

## EDA Summary

Key observations from the exploratory analysis:

- The dataset is extremely imbalanced, with fraud representing a very small fraction of transactions.
- Most features are anonymized, requiring model-driven interpretation rather than domain intuition.
- Transaction amount and time show different distributions between fraudulent and legitimate transactions.
- Due to class imbalance, evaluation metrics beyond accuracy will be required in the modeling phase.

The next step is to build classification models that properly handle imbalance and evaluate them using appropriate metrics.
